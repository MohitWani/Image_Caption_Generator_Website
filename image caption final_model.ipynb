```python
%%writefile streamlit_app.py 
import streamlit as st 
import matplotlib.pyplot as plt
from promptcap import PromptCap
import torch
import warnings
warnings.filterwarnings('ignore')

def image_caption(image_path):
  model = PromptCap("vqascore/promptcap-coco-vqa")  

  if torch.cuda.is_available():
    model.cuda()

  image = plt.imread(image_path)
  image = plt.imshow(image)
  plt.show(image)

  prompt = "please describe this image according to the given question: what is in the picture?"
  

  print(model.caption(prompt, image_path))

st.title("Image Captioning")
st.write("Upload an image and the app will generate a caption.")

uploaded_file = st.file_uploader("Choose an image...", type="png")

if uploaded_file is not None:
    st.image(uploaded_file, caption='Uploaded Image.')
    
    caption = image_caption(uploaded_file)
    st.write(caption)

```
